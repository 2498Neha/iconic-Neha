```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x1)
x=pd.DataFrame(x,columns=x1.columns)
x
```

# Handling class imbalance problem by oversampling the minority class

```python
! pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
SM=SMOTE()
x_under,y_under =SM.fit_resample(x,y)
```

```python
y_under.value_counts()
```

# Creating Train test split


```python
x=x_under
x.shape
```

```python
y=y_under
y.shape
```

```python
y=y.values.reshape(-1,1)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
maxAccu=0
maxRS=0
for i in range(1,100):
    x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=.30,random_state=i)
    lg=LogisticRegression()
    lg.fit(x_train,y_train)
    predrf=lg.predict(x_test)
    acc=accuracy_score(y_test,predrf)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print('best accuracy is',maxAccu,'on random_state',maxRS)        
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=61)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Finding Best Algorithm Using Multiple Models

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
pred=lg.predict(x_test)
print('accuracy_score:',)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
predg=gnb.predict(x_test)
print('accuracy_score:',)
print(accuracy_score(y_test,predg))
print(confusion_matrix(y_test,predg))
print(classification_report(y_test,predg))
```

```python
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
print('accuracy_score:',)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
ad=AdaBoostClassifier(n_estimators=100)
ad.fit(x_train,y_train)
pred=ad.predict(x_test)
print('accuracy_score:',)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
pred=dtc.predict(x_test)
print('accuracy_score:',)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

We are getting highest accuracy with RandomForestClassifier(95%) but it can be due to overfitting also so we check cross validation scores
