# Removing The Outliers

WE are using zscore method to remove the outliers.

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
z[39][3]
```

```python
z[5986][3]
```

```python
df_new=df[(z<3).all(axis=1)]
df_new
```

```python
df.shape
```

```python
df_new.shape
```

We can see 115 rows are deleted as outliers.
