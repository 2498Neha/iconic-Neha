# Handling class imbalance problem by using oversampling the minority class.

```python
! pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
SM=SMOTE()
x_over,y_over=SM.fit_resample(df1,y1)
```

```python
y_over.value_counts()
```

So from the above output we can see that we have balanced the class.

# Creating Train test split

```python
x=x_over
x.shape
```

```python
y=y_over
y.shape
```

```python
y=y.values.reshape(-1,1)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=43)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Finding Best Algorithm using multiple models.

```python
lg=LogisticRegression()
dtc=DecisionTreeClassifier()
knn=KNeighborsClassifier()
gnb=GaussianNB()
rf=RandomForestClassifier()
ad=AdaBoostClassifier()
svc=SVC()
```

```python
model=[lg,dtc,knn,gnb,rf,ad,svc]

for m in model:
    m.fit(x_train,y_train)
    pred=m.predict(x_test)
    print('accuracy_score:',m)
    print(accuracy_score(y_test,pred))
    print(confusion_matrix(y_test,pred))
    print(classification_report(y_test,pred))
```

We are getting 85% as highest accuracy with random forest classifier, but it may be due to overfitting so we check for cross validation score.

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
for m in model:
    score=cross_val_score(m,x,y,cv=5)
    print('score:',m)
    print(score)
    print(score.mean())
```

From the cross_validation score we can see that SVC has least dofference in accuracy and 
cross_val_score so this is our best model.

# Hyper Parameter Tunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_param={'criterion':['gini','entropy']}
gd_sr=GridSearchCV(estimator=dtc,
                 param_grid=grid_param,
                 scoring='accuracy',cv=5)
gd_sr.fit(x_train,y_train)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

```python
dtc=DecisionTreeClassifier(criterion='entropy')
dtc.fit(x_train,y_train)
dtc.score(x_train,y_train)
```

```python
parameters={'kernel':['linear','rbf'],'C':[1,10]}
clf = GridSearchCV(svc,parameters)
clf.fit(x_train,y_train)
clf

print(clf.best_params_)        
```

```python
svc=SVC(kernel='linear',C=1)
svc.fit(x_train,y_train)
t=svc.score(x_train,y_train)
print(round(t,2))
```

```python
grid_params_={'n_neighbors':np.arange(1,12,2),
             'weights':['uniform','distance']}
gscv=GridSearchCV(knn,grid_params_,cv=5)
gscv.fit(x_train,y_train)

best_parameters=gscv.best_params_
print(best_parameters)
```

```python
knn=KNeighborsClassifier(weights='distance')
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
```

Now we use  best parameter in our model

```python
svc=SVC(kernel='rbf')
svc.fit(x_train,y_train)
pred=svc.predict(x_test)
print('accuracy_score:',)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

So SVC is our best model.
