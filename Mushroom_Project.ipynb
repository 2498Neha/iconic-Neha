Here we can see that our whole dataset is present in  string format. So we have to change it in integer format.

We can done  this by applying Label Encoding on the dataset.

```python
from sklearn import preprocessing
```

```python
le=preprocessing.LabelEncoder()
```

```python
dt=df.apply(le.fit_transform)
dt
```

```python
dt['class'].unique()
```

```python
dt['class']
```

Now we are having our class  and whole dataset is in  integer format.

```python
mushroom_target=pd.DataFrame(dt['class'])
```

```python
mushroom_target
```

```python
y=mushroom_target
y
```

Till here y is created

```python
mushroom_data=pd.DataFrame(dt)
```

```python
x=mushroom_data
x
```

Now we have created both x and y for train_test_split

```python
x.shape
```

```python
y.shape
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.25,random_state=53)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
train_y.shape
```

```python
test_y.shape
```

Now we use different models for prediction.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

```

To save time we use a for loop to run all the models.

```python
model=[DecisionTreeClassifier(),SVC(),KNeighborsClassifier(),GaussianNB(),LogisticRegression()]

for m in model:
    m.fit(train_x,train_y)
    m.score(train_x,train_y)
    predm=m.predict(test_x)
    print('Accuracy score of ',m,'is:')
    print(accuracy_score(test_y,predm))
    print(confusion_matrix(test_y,predm))
    print(classification_report(test_y,predm))
    print('\n')
          
```

WE can see that all models except KNeighborsClassifier() has 100 % accuracy .
So now we will check for cross validation and get our best model.

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
gnbscores=cross_val_score(GaussianNB(),x,y,cv=5)
print(gnbscores)
print(gnbscores.mean(),gnbscores.std())
```

```python
svcscores =cross_val_score(SVC(),x,y,cv=5)
print(svcscores)
print(svcscores.mean(),svcscores.std())
```

```python
dtcscores =cross_val_score(DecisionTreeClassifier(),x,y,cv=5)
print(dtcscores)
print(dtcscores.mean(),dtcscores.std())
```

```python
knnscores=cross_val_score(KNeighborsClassifier(),x,y,cv=5)
print(knnscores)
print(knnscores.mean(),knnscores.std())
```

```python
lgscores=cross_val_score(LogisticRegression(),x,y,cv=5)
print(lgscores)
print(lgscores.mean(),lgscores.std())
```

By Looking  the results from cross validation we can see that we are getting 100% accuracy from GaussianNB and DecisionTreeClassifier.


So I am choosing DecisionTreeClassifier as my best model.
